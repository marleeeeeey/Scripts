```python
%matplotlib inline

import operator

import numpy as np
import matplotlib.pyplot as plt

from pandas import Series, DataFrame
import pandas as pd

import sklearn
from sklearn.preprocessing import scale
from sklearn.cross_validation import KFold
from sklearn.neighbors import KNeighborsClassifier
from sklearn.neighbors import KNeighborsRegressor
from sklearn.cross_validation import cross_val_score
from sklearn import datasets
```

```python
obj = datasets.load_boston()

# Приведение признаков к одному масштабу
# Получение матрицы, в которой каждый столбец имеет нулевое 
# среднее значение и единичное стандартное отклонение.
X = scale(obj.data)

y = obj.target
neighborsNum = 5    # количество соседей
splitsNum = 5
```

```python
# список параметров метрики - используется для настройки анализа
params = np.linspace(1.0, 10.0, num=200)


# генератор разбиений, который задает набор 
# разбиений на обучение и валидацию
kfold = KFold(X.shape[0], n_folds=splitsNum,
              shuffle=True, random_state=42)


# данный параметр добавляет в алгоритм веса, 
# зависящие от расстояния до ближайших соседей
# KNeighborsRegressor
model_knr = KNeighborsRegressor(n_neighbors=neighborsNum,                                   
                                weights='distance')


# реализация метода k-ближайших соседей
# принимает на вход число соседей для построения прогноза
model_knc = KNeighborsClassifier(n_neighbors=neighborsNum
                                 , metric='minkowski'
                                 #, p=kn_regressor
                                 # , weights=kn_regressor
                                )
# neighborsMethod.fit(X, y)
```

```python
# В качестве метрики используем среднеквадратичную ошибку
cvsArray = cross_val_score(model_knr, X=X, y=y, cv=kfold)
```

```python
cvs = cvsArray.mean()
```

```python
answers = {}

for p in params:
    kfold = KFold(X.shape[0], n_folds=splitsNum,
                  shuffle=True, random_state=42)
    
    model_knr = KNeighborsRegressor(n_neighbors=neighborsNum,                                   
                                    weights='distance',
                                    p = p)
    
    cvsArray = cross_val_score(model_knr, X=X, y=y, cv=kfold)
    cvs = cvsArray.mean()
    answers[p] = cvs
```
