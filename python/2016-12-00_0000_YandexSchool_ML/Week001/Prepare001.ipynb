```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.random.normal(loc = 1, scale = 10, size=(1000, 50))
plt.plot(X)
```

```python
m = np.mean(X, axis = 0)
std = np.std(X, axis = 0)
X_norm = ((X-m) / std)
plt.plot(X_norm)
```

```python
Z = np.array([[4, 5, 0],
             [1, 9, 3], 
             [5, 1, 1], 
             [3, 3, 3], 
             [9, 9, 9], 
             [4, 7, 1]])
plt.plot(Z)
```

```python
r = np.sum(Z, axis = 1)
plt.plot(r)
```
