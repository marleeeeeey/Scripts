```python
dataFemale = data[data['Sex'] == 'female']
femaleFullNames = dataFemale['Name']

femaleNames = []
for s in femaleFullNames:
    rawName = str(s).split(',')[1]
    rawName = rawName.replace('Mrs. ', '')
    rawName = rawName.replace('Miss. ', '')
    rawName = rawName.replace('(', '')
    rawName = rawName.replace(')', '')
    rawName = rawName.strip()    
    rawName = rawName.split(' ')[0]    
    femaleNames.append(rawName)

femaleNames
```

```python
uniqueFemaleNames = set(femaleNames)

namesCountDict = {}
for uName in uniqueFemaleNames:
    namesCountDict[uName] = femaleNames.count(uName)
    
import operator
sortedByCount = sorted(namesCountDict.items(), 
                       key=operator.itemgetter(1))

sortedByCount
```
